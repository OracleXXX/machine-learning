```python
import pandas as pd
import datetime
import numpy as np

# make a dataframe
df = pd.DataFrame()
# make a year and month column at random
df['year'] = np.random.randint(1986, 2018, 100)
df['month'] = np.random.randint(1, 12, 100)
df['day'] = np.random.randint(1, 28, 100)

# make a datetime column
df['date']= df.apply(lambda row :
                          datetime.datetime(row.year,row.month,row.day), 
                          axis=1)
```

```python
df.head()
```

```python
# make these additional columns
date_col_names = ['year', 'day', 'week', 'dayofweek', 'quarter', 'dayofyear', 'is_month_end', 'is_month_start', 
'is_quarter_end', 'is_quarter_start', 'is_year_end', 'is_year_start', 'weekday', 
'is_weekend', 'is_saturday', 'is_sunday', 'is_leap_year']

# use getattr
for col in date_col_names:
    try:
        df[col] = getattr(df.date.dt, col)
    except:
        if col == 'is_weekend':
            df[col] = getattr(df.date.dt, 'weekday').apply(lambda x: x in [5, 6])
        
        if col == 'is_saturday':
            df[col] = getattr(df.date.dt, 'weekday').apply(lambda x: x in [5])
            
        if col == 'is_sunday':
            df[col] = getattr(df.date.dt, 'weekday').apply(lambda x: x in [6])
```
